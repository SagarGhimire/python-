## Abstract:

This Notebook will answer my question about how does the public reacts in twitter when Apple and Samsung release their phones. I have used tweepy module to extract the data from twitter. The twitter accounts that I have extracted tweets from Apple's CEO Tim Cook and Samsung official twitter page. 

## Twitter Authorization using tweepy

```python
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
import tweepy

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
```

## Method to extract the tweet on specific days
As the product releases on specific day, so I want to see the data for that specific dates.

```python
def tweets(username,startdate, enddate):
    tweets = []
    tmpTweets = api.user_timeline(username)
    while ( tmpTweets[-1].created_at > startdate):
        tmpTweets = api.user_timeline(username, max_id = tmpTweets[-1].id)
        for tweet in tmpTweets:
            if tweet.created_at < endDate and tweet.created_at > startdate:
                tweets.append(tweet)
                
    return tweets
   
    
```

## Apple Release date and Samsung release date

```python
import datetime
Apple_user = 'tim_cook'
startDate = datetime.datetime(2018, 9, 12,0,0,0)
endDate =   datetime.datetime(2018, 9, 13,0,0,0)

Samsung_user = 'SamsungMobile'
startDate1 = datetime.datetime(2018, 3, 8,0,0,0)
endDate1 =   datetime.datetime(2018, 3, 10,0,0,0)

```

## Storing the data as pickled file

```python
import pickle
with open('apple.pkl', 'wb') as f:
    pickle.dump((tweets(Apple_user,startDate,endDate)), f)

with open('samsung.pkl', 'wb') as f:
    pickle.dump(tweets(Samsung_user,startDate1,endDate1), f)
with open('apple.pkl', 'rb') as f:
    apple_tweets = pickle.load(f)
with open('samsung.pkl', 'rb') as f:
    samsung_tweets = pickle.load(f)
    
```

## Get the Replies using Beautiful Soup

```python

import requests
import re

r = requests.get('http://docs.python-requests.org/en/master/')

from bs4 import BeautifulSoup
#getting the apple's replies
r = requests.get('https://twitter.com/tim_cook/status/1039990054469292033')
soup = BeautifulSoup(r.text, 'html5lib')
replies = soup.find('ol', {"class": "stream-items js-navigable-stream"})
original = str(replies.text)

#samsung replies
r1 = requests.get('https://twitter.com/SamsungMobile/status/1098307228241584129')
sam_soup = BeautifulSoup(r1.text, 'html5lib')
sam_replies = sam_soup.find('ol', {"class": "stream-items js-navigable-stream"})
sam_original = str(sam_replies.text)

#storing the replies as pickled file
with open('applereplies.pkl', 'wb') as f:
    pickle.dump((' '.join(original.split())), f)

with open('applereplies.pkl', 'rb') as f:
    apple_replies = pickle.load(f)

with open('samsungreplies.pkl', 'wb') as f:
    pickle.dump((' '.join(sam_original.split())), f)

with open('samsungreplies.pkl', 'rb') as f:
    samsung_replies = pickle.load(f)

```

## Sentiment Analysis on the Replies for product tweets

```python
import nltk
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()

def plottable_sentiments(tweets):
    sents = [analyzer.polarity_scores(tweets)]
    pos_x = [s['pos'] for s in sents]
    neu_y = [s['neu'] for s in sents]
    neg_z = [s['neg'] for s in sents]
    return pos_x, neu_y, neg_z
apx,apy,apz = plottable_sentiments(apple_replies)
print(apx,apy,apz)
spx,spy,spz = plottable_sentiments(samsung_replies)
print(spx,spy,spz)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(apx, apy, apz, c='r', marker='o')
ax.scatter(spx, spy, spz, c='g', marker='^')
ax.set_xlabel('Positive Score')
ax.set_ylabel('Neutral Score')
ax.set_zlabel('Negative Score')
plt.show()
```

#### This looks like, the Samsung's product replies is more positive and less neutral whereas the Apple one is more negative and neutral value.

```python

import string
#removing the unicode character
def UnicodeFilter(var):
    temp = var
    temp = temp.replace(chr(0x2019), "'")
    temp = temp.replace(chr(0x2026), "")
    for x in range(127381, 129305):
        temp = temp.replace(chr(x), "")
    return (temp)
#sentiment analysis for the tweets
def plottable_sentiments1(tweets):
    texts = [tweet.text for tweet in tweets]
    texts1 = UnicodeFilter(''.join(texts))
    sents = [analyzer.polarity_scores(texts1)]
  
    pos_x = [s['pos'] for s in sents]
    neu_y = [s['neu'] for s in sents]
    neg_z = [s['neg'] for s in sents]
    return pos_x, neu_y, neg_z

appx,appy,appz = plottable_sentiments1(tim_tweets)
sapx,sapy,sapz = plottable_sentiments1(Samsungs_tweets)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(appx, appy, appz, c='r', marker='o')
ax.scatter(sapx, sapy, sapz, c='g', marker='^')
ax.set_xlabel('Positive Score')
ax.set_ylabel('Neutral Score')
ax.set_zlabel('Negative Score')
plt.show()
```

#### This looks like as a whole Samsung's tweet is more neutral whereas Apple's tweet is more negative and neutral.

### Lexical Diversity

```python
lexemes = []
x = []

def lexical_diversity(tokens):
    return len(set(tokens)) / len(tokens)

def plotting(tweets):
    lexemes1 =[]
    for tweet in tweets:
        lexemes1.extend(tweet.text)
    return lexical_diversity(lexemes1)

for tweet in samsung_tweets:
    x.extend(tweet.text.split())
print(lexical_diversity(x))

for tweet in apple_tweets:
    lexemes.extend(tweet.text.split())
print(lexical_diversity(lexemes))
```

```python
# Plotting the lexical diversity
df1 = pd.DataFrame({'Account':['Apple', 'Samsung'], 'Lexical_Diversity':[plotting(apple_tweets), plotting(samsung_tweets)]})
ax1 = df1.plot.bar(x='Account', y='Lexical_Diversity', rot=0)
```
